Вісі автоматично розширюються до розмірів даних. Надати графіку більше місця можна за допомогою функції `axis`, яка змінює поширення кожної вісі `[xmin, xmax, ymin, ymax]`

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7])
plt.show()
```

Можна також побудувати графік математичних функцій.

Використовуючи NumPy функцію `linspace`, створимо масив `x`, який містить $500$ чисел з плаваючою точкою (floats), в районі від $-2$ до $2$. Потім створимо другий масив `y`, порахований як квадрат `x`

```python
import numpy as np

x = np.linspace(-2, 2, 500)
y = x ** 2

plt.plot(x, y)
plt.show()
```

Додамо заголовок (title), написи (labels) для x та y, а також намалюємо сітку, грід (grid)

```python
plt.plot(x, y)
plt.title("Квадратичне рівняння")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

# Стилі ліній та кольори

За замовчуванням, matplotlib малює лінію між послідовними точками

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100],
         [0, 0, 100, 100, 0, 100, 130, 100, 0])

plt.axis([-10, 110, -10, 140])
plt.show()
```

Можна надати третій аргумент, щоб змінити стиль та колір лінії. Наприклад, `"g--"` означає "зелена пунктирна лінія" ("green dashed line")

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100],
         [0, 0, 100, 100, 0, 100, 130, 100, 0],
         "g--")

plt.axis([-10, 110, -10, 140])
plt.show()
```

Можна зобразити декілька ліній на одному графіку дуже просто: треба просто надати `x1, y1, [style1], x2, y2, [style2], ...`

Наприклад:

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-",
         [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")

plt.axis([-10, 110, -10, 140])
plt.show()
```

Або просто викликавши `plot` декілька разів перед тим, як викликати `show`

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

Також можна малювати прості точки замість ліній. Ось приклад з зеленими пунктирами, червоними крапками та синіми трикутниками.

Більше стилів та кольорів можна подивититись в [офіційній документації](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x , x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
plt.show()
```

Функція побудови (plot function) повертає список (list) з `Line2D` об'єктів (один на кожну лінію). Можна встановити додаткові властивості цим лініям, як от ширина лінії, стиль пунктирів та альфа-рівень (aplha level).

Повний список властивостей є в [документації](https://matplotlib.org/stable/tutorials/introductory/pyplot.html#controlling-line-properties)

```python
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

# Підграфіки (subplots)

Фігура matplotlib може містити декілька підграфіків. Ці підграфіки організовані в сітку, грід. Щоб створити підграфік можна просто викликати функцію `subplot` і задати число рядків та стовпців в фігурі, а також індекс підграфіків, на яких треба намалювати графік (*починаючи з $1$, потім зліва направо, і зверху донизу*).

Варто зазначити, що pyplot відстежує активні на даний момент підграфіки (на які можна зіслатись викликавши `plt.gca()`), тому коли ви викликаєте функцію `plot`, вона малює *активний* підграфік

```python
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1) # 2 рядки, 2 колонки, перший підграфік = верхній лівий
plt.plot(x, x)
plt.subplot(2, 2, 2) # 2 рядки, 2 колонки, другий підграфік = верхній правий
plt.plot(x, x**2)
plt.subplot(2, 2, 3) # 2 рядки, 2 колонки, третій підграфік = нижній лівий
plt.plot(x, x**3)
plt.subplot(2, 2, 4) # 2 рядки, 2 колонки, четвертий піграфік = нижній правий
plt.plot(x, x**4)
plt.show()
```

* Варто зазначити, що `subplot(223)` - це скорочення для `subplot(2, 2, 3)`

Легко також створювати підграфіки, які розповсюджуються на декілька клітинок гріду (grid cells)

```python
plt.subplot(2, 2, 1) # 2 рядки, 2 колонки, перший підграфік = верхній лівий
plt.plot(x, x)
plt.subplot(2, 2, 2) # 2 рядки, 2 колонкт, другий підграфік = верхній правий
plt.plot(x, x**2)
plt.subplot(2, 1, 2) # 2 рядки, *1* колонка, другий підграфік = низ
plt.plot(x, x**3)
plt.show()
```

Якщо треба створити більш комплексне розміщення підграфіка, можна використати `subplot2grid` замість `subplot`. Вказується число рядків та колонок в гріді, потім позиції підграфіків в цьому гріді (top-left=(0,0)) та, за бажанням, скільки рядків та/або колонок воно охоплює.

Наприклад:

```python
plt.subplot2grid((3, 3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3, 3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3, 3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3, 3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

Якщо треба більша гнучкість у позиціонуванні підграфіків, можна скористатись відповідним [туторіалом](https://matplotlib.org/stable/tutorials/intermediate/arranging_axes.html)

# Декілька фігур

Також можливо намалювати декілька фігур. Кожна фігура може містити один або декілька підграфіків. За замовчуванням, matplotlib створює `figure(1)` автоматично. Коли ви переключаєтесь між фігурами, pyplot відстежує активну на даний момент фігуру (посилання на яку можна отримати викликавши `plt.scf()`), а активний піграфік цієї фігури стає поточним підграфіком.

```python
x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Квадрат та Куб")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")

plt.figure(1) # назад до фігури 1, поточний підграфік = 212 (нижній)
plt.plot(x, -x**3, "r:")

plt.show()
```

# Pylab *vs* Pyplot *vs* Matplotlib

Є деяка плутанина у відносинах між pylab, pyplot та matplotlib. Все просто: matplotlib - це повна бібліотека, яка містить усе, включно з pylab та pyplot.

Pyplot надає деякі інструменти для побудови графіків, включно з інтерфейсом машини стану (state-machine interface) до базової (underlying) об'єктно-орієнтованої бібліотеки побудови графіків.

Pylab - це зручний модуль, який імпортує matplotlib.pyplot та NumPy в одному полі імен (namespace). Багато прикладів з pylab, але зараз це [категорично не рекомендується](https://matplotlib.org/stable/api/index.html#module-pylab) (адже `явні` імпортування краще, ніж *неявні* -- *explicit* imports are better than *implicit* ones)

# Малювання тексту (drawinf text)

Можна викликати `text`, щоб додати текст на будь-яку локацію на графіку. Треба просто специфікувати горизонтальні та вертикальні координати та сам текст, а також опціональні додаткові аргументи. Будь-який текст в matplotlib може містити TeX вирази та рівняння. Можна переглянути [офіційну документацію](https://matplotlib.org/stable/users/explain/text/mathtext.html) для більшої кількості деталей.

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "Квадратична функція\n$y = x^2$", fontsize=20, color='blue',
         horizontalalignment='center')
plt.text(px - 0.08, py, "Гарна точка", ha="right", weight="heavy")
plt.text(px + 0.05, py - 0.3, "x = %0.2f\ny = %0.2f"%(px, py), rotation=-30,
         color='gray')

plt.show()
```

* **Нотатка**: `ha` - це скорочення для `horizontalalignment`

Для більшої кількості властивостей тексту, можна звернутись до [документації](https://matplotlib.org/stable/users/explain/text/text_props.html).


Дуже часто необхідно аннотувати елементи графіків, наприклад як гарна точка вище. Функція `annotate` робить це дуже простим: треба просто ідентифікувати локацію точки, яка вас цікавить, а також позицію тексту та опціональні додаткові аргументи для тексту та стрілочки.

```python
plt.plot(x, x**2, px, py, "ro")
plt.annotate("Гарна точка", xy=(px, py), xytext=(px-1.3,py+0.5),
             color="green", weight="heavy", fontsize=14,
             arrowprops={"facecolor":"lightgreen"})

plt.show()
```

Також можна додати контури (bounding box) для тексту, використовуючи аргумент `bbox`

```python
plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "Гарна точка", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black",
                  fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "Квадратична функція\n$y = x^2$", fontsize=20, color="black",
         ha="center", bbox=bbox_props)

plt.show()
```

Просто по криколу можна додати [xkcd](https://xkcd.com/)-стилізований графік. Треба просто будувати його всередині `with plt.xkcd() `секції

```python
with plt.xkcd():
  plt.plot(x, x**2, px, py, "ro")

  bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
  plt.text(px-0.2, py, "Beautiful point", bbox=bbox_props, ha="right")

  bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black",
                    fc="#EEEEFF", lw=5)
  plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color="black",
          ha="center", bbox=bbox_props)

  plt.show()
```

# Легенди (Legends)

Найлегший спосіб додати легенду - це встановити написи (labels) на всіх лініях, а потім просто викликати функцію `legend`

```python
x = np.linspace(-1.4, 1.4, 50)
plt.plot(x, x**2, "r--", label="Квадратична функція")
plt.plot(x, x**3, "g-", label="Кубічна функція")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

# Нелінійні масштаби (Non-linear scales)

Matplotlib підтримує нелінійні масштаби, як от логарифмічні або логіт (logit) шкали

```python
x = np.linspace(0.1, 15, 500)
y = x**3/np.exp(2*x)

plt.figure(1)
plt.plot(x, y)
plt.yscale("linear")
plt.title("лінійна")
plt.grid(True)

plt.figure(2)
plt.plot(x, y)
plt.yscale("log")
plt.title("логарифмічна")
plt.grid(True)

plt.figure(3)
plt.plot(x, y)
plt.yscale("logit")
plt.title("logit")
plt.grid(True)


plt.figure(4)
plt.plot(x, y - y.mean())
plt.yscale("symlog", linthresh=0.05)
plt.title("symlog")
plt.grid(True)

plt.show()
```

# Риски та рядки (ticks and tickers)

Вісі мають маленькі позначки (marks), які називаються "рисками" ("ticks"). Якщо бути точнішим, то "ticks" - це *локація* позначок (наприклад, (-1, 0, 1)), "лінії рисок" ("tick lines") - це маленькі лінії, намальовані на цих локаціях, "підписи рисок" ("tick labels") - це підписи, намальовані біля ліній рисок (tick lines), а "рядки" ("tickers") - це об'єкти, які можуть вирішувати, де розташовувати риски (ticks). За замовчуванняв, ticker'и виконують гарну роботу в розміщенні ~ від 5 до 8 ticks на достатній відстані один від одного.

Проте, іноді треба більше контролю (наприклад, занадто багато рисочок (ticks), як на logit графіку вище). На щастя, matplotlib надає повний контроль над рисочками. Можна навіть активовувати незначні, менші (minor) рисочки.


```python
x = np.linspace(-2, 2, 100)

plt.figure(1, figsize=(15, 10))
plt.subplot(131)
plt.plot(x, x**3)
plt.grid(True)
plt.title("Рисочки за замовчуванням")

ax = plt.subplot(132)
plt.plot(x, x**3)
ax.xaxis.set_ticks(np.arange(-2, 2, 1))
plt.grid(True)
plt.title("Вручну встановлені рисочки на вісі X")

ax = plt.subplot(133)
plt.plot(x, x**3)
plt.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom=False)
ax.xaxis.set_ticks([-2, 0, 1, 2])
ax.yaxis.set_ticks(np.arange(-5, 5, 1))
ax.yaxis.set_ticklabels(["min", -4, -3, -2, -1, 0, 1, 2, 3, "max"])
plt.grid(True)
plt.title("Вручну встановлені рисочки та їх підписи\n(а також менші рисочки) на вісі Y")

plt.show()
```

# Полярні проекції (Polar projection)

Намалювати полярний графік так само просто, як поставити значення `"polar"` аргументу `projection` під час створення підграфіка

Cartesian_to_polar.gif

```python
radius = 1
theta = np.linspace(0, 2*np.pi*radius, 1000)

plt.subplot(111, projection="polar")
plt.plot(theta, np.sin(5*theta), "g-")
plt.plot(theta, 0.5*np.cos(20*theta), "b-")
plt.show()
```

# 3D-проекції

Побудова 3D-графіків досить прямолінійна: при створенні підграфіку, треба встановити `projection` на `"3d"`. Це повертає об'єкт 3D осей, який можна використати, щоб викликати `plot_surface`, надаючи координати $x$, $y$ та $z$, а також деякі опціональні аргументи. Для більшої кількості інформації про створення 3D-графіків, можна звернутись до [офіційного туторілала](https://matplotlib.org/stable/users/explain/toolkits/mplot3d.html)

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection="3d")
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1,
                                 cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

Інший спосіб відображення цих самиї даних - це через контурну діаграму (contour plot)

```python
plt.contourf(X, Y, Z, cmap=matplotlib.cm.coolwarm)
plt.colorbar()
plt.show()
```

# діаграма розсіювання (scatter plot)

Щоб зобразити діаграму розсіювання, можна просто надати координати $x$ та $y$ для точок

```python
np.random.seed(42) # щоб зробити вихідні дані цього блокнота відтворюваними

x, y = np.random.rand(2, 100)
plt.scatter(x, y)
plt.show()
```

Також можна опціонально надати масштаб кожній точці

```python
x, y, scale = np.random.rand(3, 100)
scale = 500 * scale ** 5
plt.scatter(x, y, s=scale)
plt.show()
```

І, як зазвичай, є велика кількість різних аргументів, які можна надати. Наприклад, кольори заливки та країв, а також альфа-рівень

```python
for color in ['red', 'green', 'blue']:
  n = 100
  x, y = np.random.rand(2, n)
  scale = 500.0 * np.random.rand(n) ** 5
  plt.scatter(x, y, s=scale, c=color, alpha=0.3, edgecolors="blue")

plt.grid(True)

plt.show()
```

# Лінії

Лінії можна будувати просто використавши функцію `plot`, як ми це досі робили. Проте, часто зручніше створити корисну функцію (utility function), яка будує (здавалося б) нескінченну лінію через графік, задаючи нахил та перехват (a slope and an intercept). Можна також використати функції `hlines` та `vlines`, які будують горизонтальні та вертикальні лінійні сегменти. Наприклад:

```python
def plot_line(axis, slope, intercept, **kwargs):
  xmin, xmax = axis.get_xlim()
  plt.plot([xmin, xmax],
           [xmin*slope+intercept, xmax*slope+intercept],
           **kwargs)

x = np.random.randn(1000)
y = 0.5*x + 5 + np.random.randn(1000) * 2
plt.axis([-2.5, 2.5, -5, 15])
plt.scatter(x, y, alpha=0.2)
plt.plot(1, 0, "ro")
plt.vlines(1, -5, 0, color="red")
plt.hlines(0, -2.5, 1, color="red")
plot_line(axis=plt.gca(), slope=0.5, intercept=5, color="magenta")
plt.grid(True)
plt.show()
```

# Зображення

Читання, генерування та побудова зображень в matplotlib є досить прямолінійним.

Для того, щоб прочитати зображення, можна просто імпортувати модуль `matplotlib.image` та викликати функцію `imread`, давши їй ім'я файлу (або файловий об'єкт). Воно поверне дані зображення як масив NumPy. Спробуємо це з зображенням `my_square_function.png`, яке ми зберегли раніше.

```python
import matplotlib.image as mpimg

img = mpimg.imread("my_square_function.png")
print(img.shape, img.dtype)
```

Ми завантажили зображення $480$x$640$. Кожен піксель представлено 4-елементним масивом: червоний, зелений, синій та альфа рівні, збережені як 32-бітні floats між 0 та 1. Тепер треба просто викликати `imshow`

```python
plt.imshow(img)
plt.show()
```

Та-даам! Якщо при відображенні зображення треба приховати вісі:

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

Під капотом, `imread()` використовує Python Image Library (PIL). Документація Matplotlib зараз рекомендує використовувати PIL безпосередньо:

```python
import PIL

img = np.asarray(PIL.Image.open("my_square_function.png"))
print(img.shape, img.dtype)
```

Варто зазначити, що масив тепер містить unsigned 8-бітові інтеджери (від 0 до 255). Але це не проблема, бо plt.imshow() також підтримує цей формат

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

Можна також згенерувати зображення з нуля:

```python
img = np.arange(100*100).reshape(100, 100)
print(img)
plt.imshow(img)
plt.show()
```

Оскільки ми не надали RGB рівнів, функція imshow автоматично відображає (maps) значення, як кольоровий градієнт. За замовчуванням, кольоровий градієнт іде від синього (для низьких значень) до жовтого (для високих значень), але можна обрати й інше кольорову мапу, гамму (map).

Наприклад:

```python
plt.imshow(img, cmap="hot")
plt.show()
```

Можна також згенерувати безпосередньо зображення RGB

```python
img = np.empty((20, 30, 3))
img[:, :10] = [0, 0, 0.6]
img[:, 10:20] = [1, 1, 1]
img[:, 20:] = [0.6, 0, 0]
plt.imshow(img)
plt.show()
```

Оскільки масив `img` просто маленький ($20$x$30$), коли функція `imshow` показує його, вона вирощує зображення до розмірів фігури. Уявіть розтягування оригінального зображення, залишаючи прогалини між оригінальними пікселями. Як `imshow` заповнює пусті місця? За замовчуванням, вона просто зафарбовує кожен пустий піксель, використовуючи кольори найближчих не пустих пікселів. Ця техніка може призвести до піксельованих (pixelated) зображень. Можна використовувати й інший метод інтерполяції (interpolation), як от [білінійна інтерполяція](https://en.wikipedia.org/wiki/Bilinear_interpolation) (bilinear interpolation), щоб заповнювати пусті місця. Це призводить до розмитих (blurry) країв, що може бути кращим в деяких випадках.

```python
plt.imshow(img, interpolation="bilinear")
plt.show()
```

# Анімації

Попри те, що matplotlib в більшості використовується для генерування зображень, він також може показувати анімації. Спершу, треба імпортувати` matplotlib.animation`

```python
import matplotlib.animation as animation
```

В наступному прикладі, ми починаємо зі створення точок даних, потім створюємо пустий графік, визначаємо функцію оновлення (update function), яка буде викликатись на кожній ітерації анімації. І, нарешті, ми додаємо анімацію на графік, створюючи екземпляр (instance) `FuncAnimation`.

Конструктор `FuncAnimation` приймає фігуру, функцію оновлення та опціональні аргументи. Ми визначаємо, що хочемо анімацію, довжиною 50 кадрів (frames) з 100 мілісекундами затримки між кожним кадром. На кожній ітерації, `FuncAnimation` викликає нашу функцію оновлення і пропускає їй номер кадру num (від 0 до 49 в нашому випадку), а потім додаткові аргументи, які ми визначаємо з `fargs`.

Наша функція оновлення просто встановлює дані лінії, як перші `num` точок даних (тому дані малюються поступово), і просто по приколу додамо маленьке випадкове число для кожної точки даних, щоб здавалось, ніби лінія хитається.

```python
x = np.linspace(-1, 1, 100)
y = np.sin(x**2*25)
data = np.array([x, y])

fig = plt.figure()
line, = plt.plot([], [], "r-") # починаємо з пустим графіком
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.plot([-0.5, 0.5], [0, 0], "b-", [0, 0], [-0.5, 0.5], "b-", 0, 0, "ro")
plt.grid(True)
plt.title("Пречудова анімація")

# цю функцію буде викликано на кожній ітерації
def update_line(num, data, line):
  # ми будуємо лише перші `num` точок даних
  line.set_data(data[..., :num] + np.random.rand(2, num) / 25)
  return line,

line_ani = animation.FuncAnimation(fig, update_line, frames=50,
                                   fargs=(data, line), interval=100)
plt.close() # викликаємо call() щоб уникнути відображення статичного графіка
```

Тепер відобразимо нашу анімацію. Одна з опцій - це конвертувати її в код HTML5 (використовуючи таг `<video>`) та рендерити цей код використовуючи `IPython.display.HTML`:

```python
from IPython.display import HTML

HTML(line_ani.to_html5_video())
```

Альтернативно, можна відобразити анімацію використовуючи класний маленький HTML/Javascript віджет:

```python
HTML(line_ani.to_jshtml())
```

Можна конфігурувати Matplotlib так, щоб він використовував цей віджет для рендеру анімацій за замовчуванням:

```python
matplotlib.rc('animation', html='jshtml')
```

Після цього, не буде навіть потрібно більше використовувати `IPython.display.HTML`:

```python
animation.FuncAnimation(fig, update_line, frames=50, fargs=(data, line),
                        interval=100)
```

# Збереження анімацій у відео-файли

Matplotlib покладається на сторонні бібліотеки (3rd-party libraries), як от [FFMPEG](https://www.ffmpeg.org/) або [ImageMagick](https://imagemagick.org/). В наступному прикладі буде використано FFMPEG, тому треба спочатку впевнитись, що його встановлено. Щоб зберегти анімацію у GIF-форматі, треба буде ImageMagick

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me', bitrate=1800))
line_ani.save('my_wiggly_animation.mp4', writer=writer)
```
